```python
# We usually want to group things with strings.
# ...so we want to group by department, or salary. (I (Jerry), guess the instructor specivided salary and dept because they have object dtypes.)
```

```python
df.groupby('Departments')['left'].sum().plot()
```

```python
#So we see that the native visual is a line chart, but we can change the chart.
```

```python
df.groupby('Departments')['left'].sum().plot(kind='pie')
```

```python
df.groupby('Departments')['left'].sum().plot(kind='bar')
```

```python
df.groupby('Departments')['left'].sum().plot(kind='barh')
```

```python
#Remember the name of the variable we assigned or saved matplotlib to is plt.
#So, with that variable, we can modify the size of the figure - the figure is the same as the diagram. 

#So in the following order, we have (x,y)
```

```python
plt.figure(figsize=(10,5))

df.groupby('Departments')['left'].sum().plot(kind='barh')
```

```python
#Notice the difference in the size of the figure.
```

```python
#We can even make the style of figure more interesting.
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(10,5))

df.groupby('Departments')['left'].sum().plot(kind='barh')
```
